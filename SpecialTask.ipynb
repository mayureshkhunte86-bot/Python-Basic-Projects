```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
print("First 5 Records:")
print(df.head())
```

```python
print("\nShape:")
print(df.shape)
```

```python
print("\nColumns:")
print(df.columns)
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
print("\nDataset Info:")
print(df.info())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Species', data=df)
plt.title("Species Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['SepalLengthCm'], bins=15)
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.scatterplot(
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species',
    data=df
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y='PetalLengthCm', data=df)
plt.title("Petal Length Box Plot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Species'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)
plt.title("Species Distribution")
plt.ylabel("")
plt.show()
```
