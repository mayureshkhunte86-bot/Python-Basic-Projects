```python
print("Statistical Summary:")
print(df.describe())

# Minimum values
print("\nMinimum Values:")
print(df[['price', 'mileage', 'year']].min())

# Maximum values
print("\nMaximum Values:")
print(df[['price', 'mileage', 'year']].max())

# Mean values
print("\nMean Values:")
print(df[['price', 'mileage', 'year']].mean())

# Median values
print("\nMedian Values:")
print(df[['price', 'mileage', 'year']].median())
```

```python
# ============================
# Mini Project - Ford Car Dataset
# Complete Preprocessing Pipeline
# ============================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# -----------------------------
# 1. Load Dataset
# -----------------------------
df = pd.read_csv("ford_car_dataset.csv")

print("First 5 Rows:")
print(df.head())

print("\nDataset Shape:")
print(df.shape)

# -----------------------------
# 2. Handle Missing Values
# -----------------------------
print("\nMissing Values:")
print(df.isnull().sum())

# Remove missing values (if any)
df = df.dropna()

# -----------------------------
# 3. Remove Duplicate Rows
# -----------------------------
print("\nDuplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After Removal:", df.duplicated().sum())

# -----------------------------
# 4. Exploratory Data Analysis
# -----------------------------

# Histograms
numeric_columns = ['price', 'mileage', 'year', 'engineSize', 'mpg']

for col in numeric_columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# Count Plots
categorical_columns = ['model', 'transmission', 'fuelType']

for col in categorical_columns:
    plt.figure(figsize=(8,5))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Count Plot of {col}')
    plt.show()

# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# 5. Identify Features
# -----------------------------
X = df.drop('price', axis=1)
y = df['price']

print("\nInput Features:")
print(X.columns)

print("\nTarget Feature:")
print(y.name)

# -----------------------------
# 6. Encode Categorical Variables
# -----------------------------
X = pd.get_dummies(X, drop_first=True)

print("\nShape After Encoding:")
print(X.shape)

# -----------------------------
# 7. Scale Numeric Features
# -----------------------------
scaler = StandardScaler()

numeric_features = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

X[numeric_features] = scaler.fit_transform(X[numeric_features])

print("\nScaled Data (First 5 Rows):")
print(X.head())

print("\nPreprocessing Completed Successfully!")
```
